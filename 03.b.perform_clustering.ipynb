# Reproduce Figure 6 Clustering Results
You will need the following files:
- `file_path`: ./results/example/attn_output_residual_mlp_pt/creation_layer_info_dict_deepseek-14b_n_{n_gen}_other_{target_layer_idx}.pt (for numerous layers from `03.a.collect_clustering_data.ipynb`)


```python
import torch
import matplotlib.pyplot as plt
from transformers import AutoModelForCausalLM, AutoTokenizer
import os 

model_name = "deepseek-ai/DeepSeek-R1-Distill-Qwen-14B"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
    device_map="auto",
    attn_implementation="sdpa",
    cache_dir="/data/models", 
)
model.eval()
```

```python
# ================================================================================================================
# T-SNE
# ================================================================================================================
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

# parameters
n_gen = 10
layer_range = range(17, 23)
plot_list = ['residual', 'attn_output', 'mlp_input']
use_centering = True

# check if the data files exist
for target_layer_idx in layer_range:
    file_path = f"./results/example/attn_output_residual_mlp_pt/creation_layer_info_dict_deepseek-14b_n_{n_gen}_other_{target_layer_idx}.pt"
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Data file not found: {file_path}")

# set up the figure grid
num_layers = len(layer_range)
num_cols = len(plot_list)
fig, axes = plt.subplots(num_layers, num_cols, figsize=(6 * num_cols, 5 * num_layers))
if num_layers == 1:
    axes = np.expand_dims(axes, axis=0)

for row_idx, target_layer_idx in enumerate(layer_range):
    print(f"Layer {target_layer_idx} analysis:")
    save_dict = torch.load(
        f"./results/example/attn_output_residual_mlp_pt/creation_layer_info_dict_deepseek-14b_n_{n_gen}_other_{target_layer_idx}.pt",
        map_location="cpu"
    )

    # --- Load data ---
    # mlp_input_list = save_dict["cliff"]["mlp_input"].to(torch.float32)
    attn_output_list = save_dict["cliff"]["attn_output"].to(torch.float32)
    residual_list = save_dict["cliff"]["residual"].to(torch.float32)
    mlp_input_list = model.model.layers[target_layer_idx].mlp(save_dict["cliff"]["mlp_input"].to('cuda')).to(torch.float32)
    

    # other_mlp_input_list = save_dict["other"]["mlp_input"].to(torch.float32)
    other_attn_output_list = save_dict["other"]["attn_output"].to(torch.float32)
    other_residual_list = save_dict["other"]["residual"].to(torch.float32)
    other_mlp_input_list = model.model.layers[target_layer_idx].mlp(save_dict["other"]["mlp_input"].to('cuda')).to(torch.float32)

    print(mlp_input_list.shape, other_mlp_input_list.shape)

    # --- Random sampling for balance ---
    torch.manual_seed(42)
    other_sample_num = mlp_input_list.size(0)
    rand_idx = torch.randperm(other_mlp_input_list.size(0))[:other_sample_num]
    sample_other_mlp_input_list = other_mlp_input_list[rand_idx]
    sample_other_attn_output_list = other_attn_output_list[rand_idx]
    sample_other_residual_list = other_residual_list[rand_idx]

    # --- Combine data ---
    mask = torch.cat((
        torch.ones(mlp_input_list.shape[0], dtype=torch.bool),
        torch.zeros(sample_other_mlp_input_list.shape[0], dtype=torch.bool)
    ))

    total_attn = torch.cat((attn_output_list, sample_other_attn_output_list), dim=0)
    total_residual = torch.cat((residual_list, sample_other_residual_list), dim=0)
    total_mlp = torch.cat((mlp_input_list, sample_other_mlp_input_list), dim=0)

    if use_centering:
        total_attn -= total_attn.mean(dim=0, keepdim=True)
        total_residual -= total_residual.mean(dim=0, keepdim=True)
        total_mlp -= total_mlp.mean(dim=0, keepdim=True)

    data_map = {
        "residual": total_residual,
        "attn_output": total_attn,
        "mlp_input": total_mlp,
    }

    # --- loop through state types (columns) ---
    for col_idx, state_name in enumerate(plot_list):
        x = data_map[state_name]
        x_np = x.detach().cpu().numpy()

        # --- Step 1: K-Means clustering (optional, for diagnostics) ---
        kmeans = KMeans(n_clusters=2, random_state=42)
        _ = kmeans.fit_predict(x_np)

        # --- Step 2: t-SNE embedding ---
        tsne = TSNE(n_components=2, perplexity=30, random_state=42)
        x_tsne = tsne.fit_transform(x_np)

        # s_c = "#f57c6e"
        # r_c = "#fbe79e"

        s_c = "#ff6b35"
        r_c = "#1a659e"

        # --- Step 3: Plot ---
        ax = axes[row_idx, col_idx]
        if "mlp_input" in state_name:
            ax.scatter(x_tsne[np.array(mask) == 0, 0], x_tsne[np.array(mask) == 0, 1],
                    label=r"$y_{t!=sink}^{mlp}$", s=25, alpha=0.7, c=s_c)
            ax.scatter(x_tsne[np.array(mask) == 1, 0], x_tsne[np.array(mask) == 1, 1],
                    label="$y_{t=sink}^{mlp}$", s=25, alpha=0.7, c=r_c)
            ax.legend(fontsize=16,loc="upper right")
        elif "attn_output" in state_name:
            ax.scatter(x_tsne[np.array(mask) == 0, 0], x_tsne[np.array(mask) == 0, 1],
                    label=r"$y_{t!=sink}^{attn}$", s=25, alpha=0.7, c=s_c)
            ax.scatter(x_tsne[np.array(mask) == 1, 0], x_tsne[np.array(mask) == 1, 1],
                    label="$y_{t=sink}^{attn}$", s=25, alpha=0.7, c=r_c)
            ax.legend(fontsize=16,loc="upper right")
        elif "residual" in state_name:
            ax.scatter(x_tsne[np.array(mask) == 0, 0], x_tsne[np.array(mask) == 0, 1],
                    label=r"$h_{t!=sink}$", s=25, alpha=0.7, c=s_c)
            ax.scatter(x_tsne[np.array(mask) == 1, 0], x_tsne[np.array(mask) == 1, 1],
                    label=r"$h_{t=sink}$", s=25, alpha=0.7, c=r_c)

            ax.legend(fontsize=16,loc="upper right")
                       
        # ax.set_xlabel("t-SNE 1")
        # ax.set_ylabel("t-SNE 2")
        if "attn_output" in state_name:
            # ax.set_title(f"Layer {target_layer_idx} - MLP Output")
            ax.set_title(f"Layer {target_layer_idx}", fontsize=16)
        # else:
        #     ax.set_title(f"Layer {target_layer_idx}")
        #     # ax.set_title(f"Layer {target_layer_idx} - {state_name}")
        ax.grid(True, alpha=0.3)

    # fig.text(
    #     0.5,
    #     1 - (row_idx + 0.5) / 2,
    #     f"Layer {target_layer_idx}",
    #     ha="center",
    #     va="center",
    #     fontsize=16,
    #     fontweight="bold"
    # )

# shared labels
fig.supxlabel("t-SNE 1")
fig.supylabel("t-SNE 2")

# --- Finalize plot ---
# axes[-1, -1].legend()
plt.tight_layout()
# save as pdf
plt.savefig("tsne_plots_full.pdf", bbox_inches='tight', dpi=300)
plt.show()
```
